```python
import matplotlib.pyplot as plt
import numpy as np

# Generate normal distribution plot
x = np.linspace(-4, 4, 400)
mu = 0
sigma = 1
y = 1/(np.sqrt(2*np.pi)*sigma) * np.exp(-(x-mu)**2/(2*sigma**2))

# 畫圖
plt.figure()
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("PDF")
plt.title("Standard Normal Distribution")
plt.grid(True)
```

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np

# Reuse the x-values from the previous plot
x = np.linspace(-4, 4, 400)

# Calculate the CDF using scipy.stats.norm.cdf
# For a standard normal distribution (mu=0, sigma=1)
cdf_values = norm.cdf(x, loc=0, scale=1)

# 畫圖
plt.figure()
plt.plot(x, cdf_values)
plt.xlabel("x")
plt.ylabel("CDF")
plt.title("Standard Normal Distribution CDF")
plt.grid(True)
plt.show()
```

This code block calculates the Cumulative Distribution Function (CDF) for a standard normal distribution (mean = 0, standard deviation = 1) using `scipy.stats.norm.cdf`. It then plots the CDF values against the `x` values.
