# 求底下矩陣的特徵值與特徵向量

$$ \left[\begin{matrix}3 & 1 & 0\\1 & 2 & 1\\0 & 1 & 3\end{matrix}\right]$$

```python

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# === 1. 定義矩陣 ===
A = np.array([
    [3, 1, 0],
    [1, 2, 1],
    [0, 1, 3]
])

print("原始矩陣 A：\n", A)

# === 2. 計算特徵值與特徵向量 ===
eigenvalues, eigenvectors = np.linalg.eig(A)

print("\n特徵值 (Eigenvalues)：\n", eigenvalues)
print("\n特徵向量 (Eigenvectors)：\n", eigenvectors)

# === 3. 視覺化 ===
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# 原點
origin = np.zeros((3, 3))

# 將每個特徵向量以箭頭繪出
colors = ['r', 'g', 'b']
for i in range(3):
    ax.quiver(
        origin[0, i], origin[1, i], origin[2, i],
        eigenvectors[0, i], eigenvectors[1, i], eigenvectors[2, i],
        color=colors[i],
        label=f'λ{i+1} = {eigenvalues[i]:.2f}'
    )

# 座標軸與格式
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3x3 矩陣的特徵向量視覺化')
ax.legend()

plt.show()
```

逐次超鬆弛（successive over-relaxation，SOR）迭代法是高斯-賽德爾迭代的一種變體，用於求解線性方程組。類似方法也可用於任何緩慢收斂的迭代過程。

```python
import numpy as np
from scipy import linalg

def sor_solver(A, b, omega, initial_guess, convergence_criteria):
    """
    This is an implementation of the pseudo-code provided in the Wikipedia article.
    Arguments:
        A: nxn numpy matrix.
        b: n dimensional numpy vector.
        omega: relaxation factor.
        initial_guess: An initial solution guess for the solver to start with.
        convergence_criteria: The maximum discrepancy acceptable to regard the current solution as fitting.
    Returns:
        phi: solution vector of dimension n.
    """
    step = 0
    phi = initial_guess[:]
    residual = linalg.norm(A @ phi - b)  # Initial residual
    while residual > convergence_criteria:
        for i in range(A.shape[0]):
            sigma = 0
            for j in range(A.shape[1]):
                if j != i:
                    sigma += A[i, j] * phi[j]
            phi[i] = (1 - omega) * phi[i] + (omega / A[i, i]) * (b[i] - sigma)
        residual = linalg.norm(A @ phi - b)
        step += 1
        print("Step {} Residual: {:10.6g}".format(step, residual))
    return phi

# An example case that mirrors the one in the Wikipedia article
residual_convergence = 1e-8
omega = 0.5  # Relaxation factor

A = np.array([[4, -1, -6, 0],
              [-5, -4, 10, 8],
              [0, 9, 4, -2],
              [1, 0, -7, 5]])

b = np.array([2, 21, -12, -6])

initial_guess = np.zeros(4)

phi = sor_solver(A, b, omega, initial_guess, residual_convergence)
print(phi)
```
