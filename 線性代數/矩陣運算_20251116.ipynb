# matrix multiplicationä½œæ¥­101:




$$
\begin{pmatrix}
1 & 2 \\
3 & 4 \\
5 & 6
\end{pmatrix}
\times
\begin{pmatrix}
7 & 8 & 9 \\
10 & 11 & 12
\end{pmatrix}
=
\begin{pmatrix}
1*7+2*10 & 1*8+2*11 & 1*9+2*12 \\
 &  &  \\
 &  &
\end{pmatrix}
=
\begin{pmatrix}
27 & 30 & 33 \\
61 & 68 & 75 \\
95 & 106 & 117
\end{pmatrix}
$$


````python
# Define the first matrix (3x2)
matrix_a = np.array([
    [1, 2],
    [3, 4],
    [5, 6]
])

# Define the second matrix (2x3)
matrix_b = np.array([
    [7, 8, 9],
    [10, 11, 12]
])

print("Matrix A:")
print(matrix_a)
print("\nMatrix B:")
print(matrix_b)

# Perform matrix multiplication
result_matrix = matrix_a @ matrix_b

print("\nResult of Matrix A * Matrix B:")
print(result_matrix)
print(f"The shape of the result matrix is: {result_matrix.shape}")
````
