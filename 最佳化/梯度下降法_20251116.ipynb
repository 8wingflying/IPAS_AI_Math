```python
import numpy as np

# 目標函數與導數
def f(x):
    return x**2 + 3*x + 2

def df(x):
    return 2*x + 3

# Gradient Descent
lr = 0.1          # 學習率
x = 5.0           # 初始值
epochs = 30       # 迭代次數

history = []

for i in range(epochs):
    grad = df(x)
    x = x - lr * grad
    history.append((i, x, f(x)))
    print(f"iter {i:02d}: x={x:.5f}, f(x)={f(x):.5f}")

print("\n最佳解約為 x =", x)
```

```python
import matplotlib.pyplot as plt

# Extract data from history
iterations = [item[0] for item in history]
x_values = [item[1] for item in history]
f_x_values = [item[2] for item in history]

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(iterations, x_values, marker='o', linestyle='-')
plt.title('X Value vs. Iteration')
plt.xlabel('Iteration')
plt.ylabel('X Value')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(iterations, f_x_values, marker='o', linestyle='-', color='red')
plt.title('f(x) Value vs. Iteration')
plt.xlabel('Iteration')
plt.ylabel('f(x) Value')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# 建立假數據
np.random.seed(0)
X = np.linspace(0, 10, 50)
y = 3 * X + 5 + np.random.randn(50)  # 真實 w=3, b=5

# 參數初始化
w = 0.0
b = 0.0
lr = 0.01
epochs = 2000

N = len(X)

for i in range(epochs):
    y_pred = w * X + b

    # 計算梯度
    dw = (2/N) * np.sum(X * (y_pred - y))
    db = (2/N) * np.sum(y_pred - y)

    # 更新
    w -= lr * dw
    b -= lr * db

    if i % 200 == 0:
        loss = np.mean((y_pred - y)**2)
        print(f"iter {i:04d}: loss={loss:.4f}, w={w:.4f}, b={b:.4f}")

print("\n學到的參數：")
print("w =", w)
print("b =", b)

import matplotlib.pyplot as plt

plt.scatter(X, y, label="data")
plt.plot(X, w * X + b, color='red', label="fit")
plt.legend()
plt.show()
```
